$$
\large \textbf{Problema 03}
$$

a) Un paracaidista de $70\:kg$. salta desde un globo aerostático fijo. El coeficiente de resistencia del aire es de aproximadamente $12.5\:\frac{kg}{s}$. (tome la aceleración de la gravedad como $10\:\frac{m}{s^2}$). Utilizando la expresión exacta evalúe la velocidad del paracaidista cada $2\:s$ después del salto y calcule su velocidad límite. Grafique el resultado.\
b) Repita el gráfico con la solución discreta, tomando intervalos de $2\:s$, $1\:s$ y $0.5\:s$.\
c) Haga el mismo cálculo pero suponiendo ahora que la fuerza de rozamiento se debe expresar como $F=\:–c'\: v^2$ (tome $c' = 0.25 \:\frac{kg}{m}$).
***

```python
import numpy as np                # Librería de álgebra lineal
import matplotlib.pyplot as plt   # Librería de gráficos
plt.rc('figure', figsize=(9,6))
plt.rc('font', size=14)
```

# <u>PARTE 1</u>
### Ecuación diferencial
$$ m\:\dot{v}=m\:g-\mu\:v $$

```python
m = 70     # kg
g = 10     # m/s^2
mu = 12.5  # kg/s

vi = 0     # m/s
```

### Solución analítica
$$ v=\frac{gm}{\mu}\:\left(1-e^{-\frac{\mu}{m}t}\right) $$

```python
t = [0]   # s
v = [vi]  # m/s
dt = 2    # s
tf = 40   # s
```

```python
vM = g*m/mu
vLim = [vM]

i = dt
while t[-1]<tf:
    t.append(i)
    v.append(vM*(1-np.exp(-(mu/m)*t[-1])))
    vLim.append(vM)
    i += dt
```

### Solución numérica

```python
tD1 = [0]   # s
vD1 = [vi]  # m/s
dt1 = 2     # s
tD2 = [0]   # s
vD2 = [vi]  # m/s
dt2 = 1     # s
tD3 = [0]   # s
vD3 = [vi]  # m/s
dt3 = 0.5   # s
tF = 40     # s
```

```python
i = dt1
while tD1[-1]<tf:
    tD1.append(i)
    vD1.append(vD1[-1]+(g-mu/m*vD1[-1])*dt1)
    i += dt1
    
i = dt2
while tD2[-1]<tf:
    tD2.append(i)
    vD2.append(vD2[-1]+(g-mu/m*vD2[-1])*dt2)
    i += dt2
    
i = dt3
while tD3[-1]<tf:
    tD3.append(i)
    vD3.append(vD3[-1]+(g-mu/m*vD3[-1])*dt3)
    i += dt3
```

### Gráfico

```python
plt.plot(t, vLim, linestyle='dashed', color='gray')
plt.plot(t, v, linestyle='solid', color='green')
plt.title('Solución analítica')
plt.legend(['Vel. límite', 'Paso 2 s'])
plt.xlabel('Tiempo (s)')
plt.ylabel('Velocidad (m/s)')
plt.show()
```

```python
plt.plot(t, vLim, linestyle='dashed', color='gray')
plt.plot(tD1, vD1, linestyle='solid', color='green')
plt.plot(tD2, vD2, linestyle='solid', color='red')
plt.plot(tD3, vD3, linestyle='solid', color='blue')
plt.title('Solución numérica')
plt.legend(['Vel. límite', 'Paso 2 s', 'Paso 1 s', 'Paso 0.5 s'])
plt.xlabel('Tiempo (s)')
plt.ylabel('Velocidad (m/s)')
plt.show()
```

# <u>PARTE 2</u>
### Ecuación diferencial
$$ m\:\dot{v}=m\:g-c'\:v^2 $$

```python
m = 70     # kg
g = 10     # m/s^2
c = 0.25  # kg/m

vi = 0     # m/s
```

### Solución numérica

```python
tD1 = [0]   # s
vD1 = [vi]  # m/s
dt1 = 2     # s
tD2 = [0]   # s
vD2 = [vi]  # m/s
dt2 = 1     # s
tD3 = [0]   # s
vD3 = [vi]  # m/s
dt3 = 0.5   # s
tF = 30     # s
```

```python
i = dt1
while tD1[-1]<tF:
    tD1.append(i)
    vD1.append(vD1[-1]+(g-c/m*vD1[-1]**2)*dt1)
    i += dt1
    
i = dt2
while tD2[-1]<tF:
    tD2.append(i)
    vD2.append(vD2[-1]+(g-c/m*vD2[-1]**2)*dt2)
    i += dt2
    
i = dt3
while tD3[-1]<tF:
    tD3.append(i)
    vD3.append(vD3[-1]+(g-c/m*vD3[-1]**2)*dt3)
    i += dt3
```

### Gráfico

```python
plt.plot(tD1, vD1, linestyle='solid', color='green')
plt.plot(tD2, vD2, linestyle='solid', color='red')
plt.plot(tD3, vD3, linestyle='solid', color='blue')
plt.title('Solución numérica')
plt.legend(['Paso 2 s', 'Paso 1 s', 'Paso 0.5 s'])
plt.xlabel('Tiempo (s)')
plt.ylabel('Velocidad (m/s)')
plt.show()
```
